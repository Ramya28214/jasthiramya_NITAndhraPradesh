```python
import nest_asyncio
import uvicorn
from pyngrok import ngrok
from fastapi import FastAPI
from pydantic import BaseModel
from io import BytesIO
import requests
from PIL import Image
import pytesseract
import re
import threading

nest_asyncio.apply()
```

```python
app = FastAPI()

class DocumentRequest(BaseModel):
    document: str
```

```python
def extract_bill_items(file_bytesio):

    image = Image.open(file_bytesio)


    text = pytesseract.image_to_string(image)


    line_items = []
    for line in text.split("\n"):
        match = re.match(r"(.+?)\s+(\d+)\s+([\d.]+)\s+([\d.]+)", line)
        if match:
            name, qty, rate, amount = match.groups()
            line_items.append({
                "item_name": name.strip(),
                "item_quantity": float(qty),
                "item_rate": float(rate),
                "item_amount": float(amount)
            })

    return {
        "pagewise_line_items": [
            {
                "page_no": "1",
                "page_type": "Final Bill",
                "bill_items": line_items
            }
        ],
        "total_item_count": len(line_items)
    }
```

```python
@app.post("/extract-bill-data")
def extract_bill_data(request: DocumentRequest):

    response = requests.get(request.document)
    file_bytesio = BytesIO(response.content)


    extracted_data = extract_bill_items(file_bytesio)

    return {
        "is_success": True,
        "token_usage": {
            "total_tokens": 0,
            "input_tokens": 0,
            "output_tokens": 0
        },
        "data": extracted_data
    }
```

```python
def start_uvicorn():
    uvicorn.run(app, host="0.0.0.0", port=8000)

thread = threading.Thread(target=start_uvicorn, daemon=True)
thread.start()
```
